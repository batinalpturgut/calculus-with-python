Random polynomials in numpy
Create four random numbers (normal distribution)
Construct a 3rd order polynomial with those coefficients and plot

```python
import numpy as np
import sympy as sym
import matplotlib.pyplot as plt
```

```python
# My solution:
rand_nums = np.round(np.random.randn(4), decimals=2)  # create 4 random numbers using normal distribution

print(rand_nums)

x_domain = [-4, 4]

resolution = .1
x = np.arange(x_domain[0], x_domain[1] + resolution, resolution)  # Creating x values.

# Creating function:
y = np.zeros(len(x)) # The number of outputs is equal to the number of inputs.
for i in range(4):
    y += rand_nums[i] * x ** i

plt.plot(x, y)
plt.grid()
plt.xlim(x_domain[0], x_domain[1])
plt.xlabel('x')
plt.ylabel('y = f(x)')
plt.show()
```

If the highest power of a term is odd, despite the presence of terms with even powers, the graph of the function
looks like an odd polynomials. However, if the coefficient of $x^2$ is greater than that of $x^3$ within a restricted
domain, the graph looks more like an even polynomial function. If we increase the domain range, eventually graph is look
