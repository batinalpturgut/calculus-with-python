f(x) = $\frac{(x^2 -2x)}{(x^2 -4)}$ Implement and plot this function using sympy

```python
import matplotlib.pyplot as plt
import sympy as sym
import numpy as np
import math
```

```python
# My solution:
x_domain = [-3, 3]

s_x = sym.symbols('x')

s_y = (s_x**2 - 2*s_x) / (s_x**2 - 4)

sym.plot(s_y, (s_x, x_domain[0], x_domain[1]), ylim = [-10, 10]) # Setting x-axis and y-axis limit manually.
```

```python
# Original code:

from sympy.abc import x

# define the function in two parts:
top = x**2 - 2*x
bot = x**2 - 4

sy = top / bot
# Separating out these mathematical functions makes them easier to read.

sym.plot(sy, (x, -3, 3), ylim = [-10, 10], title = f'$y = {sym.latex(sy)}$')
```

```python
# Exercise 3, Original Code:
order = 10

xx = np.linspace(-2 * np.pi, 2 * np.pi, 100)
z = np.zeros(len(xx))

for n in range(1, order + 1):

    #create this term
    sign = (-1) ** (n + 1)
    num = xx ** (2 * n - 1)
    den = math.factorial((2*n - 1))
    this_term = sign * num/den

    # plot it
    plt.plot(xx, this_term, '--')

    # sum it onto z
    z += this_term

plt.plot(xx, z, 'k', linewidth = 3, label = f'Sum over {order} terms') # Our function
plt.plot(xx[::5], np.sin(xx[::5]), 'bo', markerfacecolor = 'w', linewidth = 3, label = 'sin(x)') # Original sin function
plt.legend()
plt.xlim([xx[0], xx[len(xx) - 1]]) # We can use -1 instead of len(xx) - 1.
plt.ylim([-5, 5])
plt.show()

# Blue dots are original sin function
```
