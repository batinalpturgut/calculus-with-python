```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sym
```

```python
# Function = x**2 + (3x)**3 - x**4
# Plot the function
# Domain = [-2, 2]

x_domain = [-2, 2]

# option 1
resolution = .1 # step
x = np.arange(x_domain[0], x_domain[1] + resolution, resolution)
print(x)
# We added resolution to x_domain[1] because x_domain[1] is normally exclusive. [-2, 2)
# Since the last element in the array is x_domain[1] - resolution, we added resolution to ensure the inclusion of x_domain[1].

# option 2
length = 41
x = np.linspace(x_domain[0], x_domain[1], length)
print(x)
# linspace() does give us inclusive lower and upper bounds.

# function
y = x**2 + 3*x**3 - x**4

print(y)

# plot the graph

plt.plot(x, y, label = '$y=x^2+3x^3-x^4$')
plt.legend() # Show the legend
plt.grid()
plt.xlim(x_domain[0], x_domain[1]) # Set x-axis limits
plt.xlabel('x')
plt.ylabel('y = f(x)')
plt.show()
```

```python
# Implement the function in sympy
# f(b) = -b^4 + 3*b^3 + b^2

# create a symbolic variable

s_beta = sym.var('beta') # s represents that the variable is a sympy variable.

# define the function
s_y = -s_beta**4 + 3*s_beta**3 + s_beta**2

sym.plot(s_y, (s_beta, x_domain[0], x_domain[1]), xlabel = 'x', ylabel = None, title = f'$f(\\beta) = {sym.latex(s_y)}$')
```

Converting sympy function into numpy function.

```python
fx = sym.lambdify(s_beta, s_y) # Input is s_beta
print(fx(2))
```

```python
n_y = fx(x) # We send the numpy array x into the function, and this provides us with a new numpy array by using the values of x within the function.

print(type(n_y))
print(n_y)

plt.plot(x, n_y)
```

```python
plt.plot(x, n_y, label = '$y=x^2+3x^3-x^4$')
plt.legend()
plt.grid()
plt.xlim(x_domain[0], x_domain[1])
plt.xlabel('x')
plt.ylabel('y = f(x)')
plt.show()
```
